# 3. Пропущенные значения

```python
data_raw.describe(include = 'all')
```

выведем пропущенные (кол-во) значения

```python
data_raw.isna().sum()
```

Заполним медианами пропущенные значения в столбцах, соответствующих числовым признакам:

```python
data_raw.fillna(data_raw.median(axis = 0), axis=0 , inplace=True)
```

```python
data_raw.isna().sum()
```

```python
data_raw.describe()
```

# 5. Нормализация количественных признаков

Выполним стандартизацию всех признаков:

```python
data_stand = (data_raw - data_raw.mean(axis = 0))/data_raw.std(axis = 0)
```

```python
data_stand.describe()
```

# 6. Отбор признаков
Для выявления заболевания будем использовать все входные признаки, кроме education

```python
data_stand.columns
```

```python
X = data_stand.drop(['TenYearCHD', 'education'], axis=1)
y = data_raw['TenYearCHD']
```

Разобьем данные на обучающую и тестовую выборки в пропорции 3:1 (75% - обучающая выборка, 25% - тестовая):

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

N_train, N_test
```

# 7. Классификация ближайших соседей

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 10)
#knn.set_params(n_neighbors=10)
knn.fit(X_train, y_train)
```

```python
knn
```

```python
y_test_predict = knn.predict(X_test)
err_test  = np.mean(y_test  != y_test_predict)
```

```python
err_test
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_predict))
```

```python
confusion_matrix([0,1,0,1,0], [0,1,0,0,0])
```
